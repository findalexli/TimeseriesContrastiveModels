```python
import numpy as np
from matplotlib import pyplot as plt
# This notebook explores conversion of 
signal_100Hz = raw_x_data[0][:, 0] # Just the x-axis
from scipy import signal

# Generate a sample signal with a frequency of 100Hz
sample_rate = 100
time = np.arange(0, 10, 1/sample_rate)  # 10 seconds of signal


# Downsample the signal from 100Hz to 4Hz
desired_sample_rate = 4
downsample_factor = int(sample_rate / desired_sample_rate)
signal_4Hz = signal.decimate(signal_100Hz, downsample_factor)

# Print the lengths of original and downsampled signals
print(len(signal_100Hz))
print(len(signal_4Hz))
def plot_spectrogram(signal, sample_rate):
    plt.specgram(signal, Fs=sample_rate, detrend='mean', NFFT= 375)
    plt.xlabel('Time')
    plt.ylabel('Frequency')
    # plt.ylim(0, 4)
    plt.show()
plt.plot(time, signal_100Hz)
#plot_spectrogram(signal_4Hz, 4)
```
