```python
# Compute STFT for each axis
import matplotlib.pyplot as plt
import numpy as np
data = dataset['realworld']['x'][0]

sample_rate = 30

# Plot STFT for each axis in a separate plot
plt.figure(figsize=(10, 8))
plt.subplot(3, 1, 1)
plt.specgram(np.squeeze(data[:, 0]), Fs=sample_rate, detrend='mean', NFFT=375, cmap='viridis')
plt.title('STFT for X-axis')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')

plt.subplot(3, 1, 2)
plt.specgram(np.squeeze(data[:, 1]), Fs=sample_rate, detrend='mean', NFFT=375, cmap='viridis')
plt.title('STFT for Y-axis')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')

plt.subplot(3, 1, 3)
plt.specgram(np.squeeze(data[:, 2]), Fs=sample_rate, detrend='mean', NFFT=375, cmap='viridis')
plt.title('STFT for Z-axis')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')

plt.tight_layout()
plt.show()
```
